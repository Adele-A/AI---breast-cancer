```python
#Importing the necessary libraries

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score
```

```python
#Reading the dataset 

df_cancer_data = pd.read_csv("breast-cancer.csv")
```

```python
#Displaying the first 12 rows

df_cancer_data.head(12)
```

```python
#Displaying the last 5 rows

df_cancer_data.tail(5)
```

```python
#Printing the number of rows and columns in the dataset

rows = len(df_cancer_data.axes[0]) #Where 0 represents horizontal axis
columns = len(df_cancer_data.axes[1]) #Where 1 represents vertical axis
print("Number of rows: " + str(rows))
print("Number of columns: " + str(columns))
```

```python
#Checking for missing values and printing the result

missing_values = df_cancer_data.isna().sum()
print(df_cancer_data.isna().sum())
```

```python
#Logistic Regression
#Extracting X feature and Y target

df_cancer_data['diagnosis'] = df_cancer_data['diagnosis'].map({"B": 0, "M": 1})
X = df_cancer_data.drop(columns=['id', 'diagnosis'])
Y = df_cancer_data['diagnosis']
```

```python
#Dividing the data into train and test sets

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=42)
```

```python
#Implementing the algorithm and finding first 10 values for the data and predicted values

lr = LogisticRegression(solver="saga", max_iter = 3500) #Varient of sag solver and is used for large datasets

#Training the model
lr.fit(X_train, Y_train)
Y_pred = lr.predict(X_test)

#Evaluation of the accuracy of the model
accuracy = accuracy_score(Y_test, Y_pred)
print(f"Accuracy: {accuracy: .2f}") #Rounded to 2 decimal points
print(f"Actual data:    {Y_test[:10].values}")
print(f"Predicted values: {Y_pred[:10]}")
```

```python
#Showing result with the Confusion Matrix

cm = confusion_matrix(Y_test, Y_pred)
print("Confusion matrix: ", cm)
```

```python
#Plotting the Confusion Matrix

sns.heatmap(cm, cmap="Purples", annot=True, 
            cbar_kws={"orientation":"vertical","label":"Color Bar"},
            xticklabels=[0,1],yticklabels=[0,1])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
#Calculating and printing the Precision, Recall and F1 Score

precision = precision_score(Y_test, Y_pred)
recall = recall_score(Y_test, Y_pred)
f1 = f1_score(Y_test, Y_pred)

#Rounded to 2dp
print(f"Precision: {precision: .2f}" + " or ", precision)
print(f"Recall: {recall: .2f}"+ " or ", recall)
print(f"F1 Score: {f1: .2f}"+ " or ", f1)
```
